```python
import pandas as pd

dataset = pd.read_csv("CourseraDataset-Unclean.csv")

# printing the shape of the dataset
dataset.shape
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
# printing the shape of the dataset
dataset.shape
```

```python
dataset["Review"].fillna(dataset["Review"].mode()[0], inplace=True)
dataset["Review"] = dataset["Review"].str.extract("(\d+)", expand=False).astype(int)
```

```python
dataset["Rating"].fillna(dataset["Rating"].mode()[0], inplace=True)
```

```python
dataset["Review"].mode()[0]
```

```python
dataset.head()
```

```python
# Check if there are ratings greater than 5
has_greater_than_5 = any(dataset["Rating"] > 5)

# Check if there are ratings less than 0
has_less_than_0 = any(dataset["Rating"] < 0)

# Display the results
if has_greater_than_5:
    dataset["Rating"].fillna(dataset["Rating"].mode()[0], inplace=True)
else:
    print("There are no ratings greater than 5.")

if has_less_than_0:
    dataset["Rating"].fillna(dataset["Rating"].mode()[0], inplace=True)
else:
    print("There are no ratings less than 0.")
```

```python
# Drop specified columns
columns_to_drop = [
    "Level",
    "Duration",
    "Schedule",
    "Skill gain",
    "Modules",
    "Instructor",
    "Offered By",
    "Keyword",
    "Course Url",
]
df = dataset.drop(columns=columns_to_drop)

# Display the modified DataFrame
df
```
